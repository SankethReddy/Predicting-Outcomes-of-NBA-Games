## Getting the EFF to be used in the model

```python
def getEFFUsedInModel(game_number_player, eff_prev_game, eff_prev_season, eff_prev_five):
    if game_number_player == 1:
        if pd.isna(eff_prev_season):
            return 0
        else:
            return eff_prev_season
    else:
        if pd.isna(eff_prev_five):
            return eff_prev_game
        else:
            if game_number_player <= 5:
                return eff_prev_game
            else:
                return eff_prev_five

df['EFF_USED_IN_MODEL'] = [getEFFUsedInModel(gnp,epg,eps,epf) for (gnp,epg,eps,epf) in zip(df['GAME_NUMBER_PLAYER'],
                                                                                          df['EFF_PREV_GAME'],
                                                                                          df['EFF_PREV_SEASON_AVG'],
                                                                                          df['EFF_PREV_FIVE_AVG'])]
df.head()
```

```python
def getTeamWinsColumn(home_team_id, visitor_team_id, team_id, home_team_wins):
    if team_id == home_team_id:
        return home_team_wins
    else:
        if home_team_wins == 0:
            return 1
        else:
            return 0

def getTeamMetricColumn(home_team_id, visitor_team_id, team_id, home_metric, away_metric):
    if team_id == home_team_id:
        return home_metric
    else:
        return away_metric

df['TEAM_PTS'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['PTS_home'], df['PTS_away'])]
df['TEAM_FG_PCT'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['FG_PCT_home'], df['FG_PCT_away'])]
df['TEAM_FT_PCT'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['FT_PCT_home'], df['FT_PCT_away'])]
df['TEAM_FG3_PCT'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['FG3_PCT_home'], df['FG3_PCT_away'])]
df['TEAM_AST'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['AST_home'], df['AST_away'])]
df['TEAM_REB'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['REB_home'], df['REB_away'])]
df['TEAM_G'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['G_HOME'], df['G_AWAY'])]
df['TEAM_W'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['W_HOME'], df['W_AWAY'])]
df['TEAM_L'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['L_HOME'], df['L_AWAY'])]
df['TEAM_W_PCT'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['W_PCT_HOME'], df['W_PCT_AWAY'])]
df['TEAM_HOME_RECORD'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['HOME_RECORD_HOME'], df['HOME_RECORD_AWAY'])]
df['TEAM_ROAD_RECORD'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['ROAD_RECORD_HOME'], df['ROAD_RECORD_AWAY'])]
df['TEAM_HOME_W_PCT'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['HOME_W_PCT_HOME'], df['HOME_W_PCT_AWAY'])]
df['TEAM_ROAD_W_PCT'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['ROAD_W_PCT_HOME'], df['ROAD_W_PCT_AWAY'])]
df['TEAM_RANK'] = [getTeamMetricColumn(ht,vt,ti,hm,am) for (ht,vt,ti,hm,am) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['RANK_HOME'], df['RANK_AWAY'])]
df['TEAM_WINS'] = [getTeamWinsColumn(h,v,t,hw) for (h,v,t,hw) in zip(df['HOME_TEAM_ID'], df['VISITOR_TEAM_ID'], df['TEAM_ID'], df['HOME_TEAM_WINS'])]
df.head()
```

```python
def getMetricUsedInModel(game_number_team, metric_prev_game, metric_prev_season, metric_prev_five):
    if game_number_team == 1:
        if pd.isna(metric_prev_season):
            return 0
        else:
            return metric_prev_season
    else:
        if pd.isna(metric_prev_five):
            return metric_prev_game
        else:
            if game_number_team <= 5:
                return metric_prev_game
            else:
                return metric_prev_five

game_team_grouped_df['TEAM_PTS_USED_IN_MODEL'] = [getMetricUsedInModel(gnt,mpg,mps,mpf) for (gnt,mpg,mps,mpf) in zip(game_team_grouped_df['GAME_NUMBER_TEAM'],
                                                                                                                     game_team_grouped_df['PTS_PREV_GAME'],
                                                                                                                     game_team_grouped_df['PTS_PREV_SEASON_AVG'],
                                                                                                                     game_team_grouped_df['PTS_PREV_FIVE_AVG'])]
game_team_grouped_df['TEAM_FG_PCT_USED_IN_MODEL'] = [getMetricUsedInModel(gnt,mpg,mps,mpf) for (gnt,mpg,mps,mpf) in zip(game_team_grouped_df['GAME_NUMBER_TEAM'],
                                                                                                                     game_team_grouped_df['FG_PCT_PREV_GAME'],
                                                                                                                     game_team_grouped_df['FG_PCT_PREV_SEASON_AVG'],
                                                                                                                     game_team_grouped_df['FG_PCT_PREV_FIVE_AVG'])]
game_team_grouped_df['TEAM_FT_PCT_USED_IN_MODEL'] = [getMetricUsedInModel(gnt,mpg,mps,mpf) for (gnt,mpg,mps,mpf) in zip(game_team_grouped_df['GAME_NUMBER_TEAM'],
                                                                                                                     game_team_grouped_df['FT_PCT_PREV_GAME'],
                                                                                                                     game_team_grouped_df['FT_PCT_PREV_SEASON_AVG'],
                                                                                                                     game_team_grouped_df['FT_PCT_PREV_FIVE_AVG'])]
game_team_grouped_df['TEAM_FG3_PCT_USED_IN_MODEL'] = [getMetricUsedInModel(gnt,mpg,mps,mpf) for (gnt,mpg,mps,mpf) in zip(game_team_grouped_df['GAME_NUMBER_TEAM'],
                                                                                                                     game_team_grouped_df['FG3_PCT_PREV_GAME'],
                                                                                                                     game_team_grouped_df['FG3_PCT_PREV_SEASON_AVG'],
                                                                                                                     game_team_grouped_df['FG3_PCT_PREV_FIVE_AVG'])]
game_team_grouped_df['TEAM_AST_USED_IN_MODEL'] = [getMetricUsedInModel(gnt,mpg,mps,mpf) for (gnt,mpg,mps,mpf) in zip(game_team_grouped_df['GAME_NUMBER_TEAM'],
                                                                                                                     game_team_grouped_df['AST_PREV_GAME'],
                                                                                                                     game_team_grouped_df['AST_PREV_SEASON_AVG'],
                                                                                                                     game_team_grouped_df['AST_PREV_FIVE_AVG'])]
game_team_grouped_df['TEAM_REB_USED_IN_MODEL'] = [getMetricUsedInModel(gnt,mpg,mps,mpf) for (gnt,mpg,mps,mpf) in zip(game_team_grouped_df['GAME_NUMBER_TEAM'],
                                                                                                                     game_team_grouped_df['REB_PREV_GAME'],
                                                                                                                     game_team_grouped_df['REB_PREV_SEASON_AVG'],
                                                                                                                     game_team_grouped_df['REB_PREV_FIVE_AVG'])]
game_team_grouped_df = game_team_grouped_df.rename(columns = {'TEAM_AVERAGE_EFF': 'TEAM_EFF_USED_IN_MODEL'})
game_team_grouped_df.head()
```

```python
def getEFFColumn(pts, reb, ast, stl, blk, fgm, fga, ftm, fta, to):
    eff = pts+reb+ast+stl+blk-(fga-fgm)-(fta-ftm)-to
    return eff

games_details_df = games_details_df[games_details_df['COMMENT'].isnull()]
games_details_df['EFF'] = [getEFFColumn(p,r,a,s,b,fm,fa,fm1,fa1,t) for (p,r,a,s,b,fm,fa,fm1,fa1,t) in
                   zip(games_details_df['PTS'], games_details_df['REB'], games_details_df['AST'], games_details_df['STL'],
                      games_details_df['BLK'], games_details_df['FGM'], games_details_df['FGA'], games_details_df['FTM'],
                      games_details_df['FTA'], games_details_df['TO'])]
```

## ranking.csv Analysis

```python
ranking_df = pd.read_csv('ranking.csv')
ranking_df.head()
```

```python
def getHomeRoadPct(record):
    r_lst = record.split('-')
    w = int(r_lst[0])
    l = int(r_lst[1])
    if w+l != 0:
        return np.round((w)/(w+l),3)
    else:
        return 0

ranking_df['HOME_W_PCT'] = [getHomeRoadPct(hr) for hr in ranking_df['HOME_RECORD']]
ranking_df['ROAD_W_PCT'] = [getHomeRoadPct(rr) for rr in ranking_df['ROAD_RECORD']]
ranking_df.head()
```
